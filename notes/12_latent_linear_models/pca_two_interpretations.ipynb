```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.parametrizations import orthogonal
import matplotlib.pyplot as plt
```

```python
data = np.random.multivariate_normal(mean=[0, 0, 0], cov=[[1, 0.5], [0.5, 1]], size=10000)
data = torch.from_numpy(data).float()
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
class PCA(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.ortho_linear = orthogonal(nn.Linear(1, 2))
        
    def forward(self, X):
        W = self.ortho_linear.weight
        return (W @ W.T @ X.T).T
```

```python
losses = []
pca = PCA()
opt = optim.Adam(pca.parameters(), lr=1e-3)
for i in range(10000):
    opt.zero_grad()
    loss = torch.linalg.matrix_norm(pca(data) - data)
    loss.backward()
    opt.step()
    losses.append(float(loss))
plt.plot(losses)
plt.show()
```

```python
x1, y1 = pca.ortho_linear.weight.detach().numpy()[:,0]
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(data[:,0], data[:,1], alpha=0.1)
plt.plot([0, x1], [0, y1], color="black")
plt.show()
```

```python
x1, y1 = pca.ortho_linear.weight.detach().numpy()[:,0]
x2, y2 = pca.ortho_linear.weight.detach().numpy()[:,1]
x1 * x2 + y1 * y2, x1 * x1 + y1 * y1, x2 * x2 + y2 * y2
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(data[:,0], data[:,1], alpha=0.1)
plt.plot([0, x1], [0, y1], color="black")
plt.plot([0, x2], [0, y2], color="black")
plt.show()
```
